#Dense Layer Class

```python
# performs weights and biases initialisation automatically, storing, forward pass everything is done modularly using this class

# We'll stick to inputs containing only 2 data points
# We'll randomise these inputs, to follow a non linear pattern

from nnfs.datasets import spiral_data # Secifically, we'll be generating spiral data
import numpy as np
import nnfs
nnfs.init()
import matplotlib.pyplot as plt
X,y = spiral_data(samples = 100, classes = 3) #X is a 300by2 numpy array of data points, y is a 300by1 nparray of class labels(0-2)
plt.scatter(X[:,0],X[:,1])
```

```python
# Dense layer

class Dense_Layer:
  def __init__ (self,n_inputs,n_neurons):
    self.weights = np.random.randn(n_inputs, n_neurons)
    self.biases = np.zeros((1,n_neurons))
  def forward(self,inputs):
    self.output = np.dot(inputs,self.weights)+self.biases
```

```python
dense1 = Dense_Layer(2,3)
dense1.forward(X)
print(dense1.output[:5])
```
